## General method

* In Python, we always use the package called Matplotlib to help us plot in Python
* The process will be
    1. Set the figure (fig)
    2. Set the axes (ax)
    3. Output/Save the figure

### Set the figure
* We can use `plt.figure()` to set the figure
* You may see someone will write `fig, ax = plt.figure()`, which can set the figure more detail.
* We can use `(figsize=())` to set the size of the figure

### Set the axes
* There are some common set-up about setting axes
    1. `plt.title()` => Setting the title
    2. `plt.plot()` => Plot the line plot
    3. `plt.scatter()` => Plot the scatter plot
    4. `plt.xlabel()` => Set the label of x-axis
    5. `plt.ylabel()` => Set the label of y-axis
    6. `plt.legend(loc=)` => Set the position and the title of the lengend
        * `loc='best'`
        * `loc='upper right'`
        * `loc='upper left'`
        * `loc='lower right'`
        * `loc='lower left'`
        * etc.

### Show/Save the figure
* `plt.show()` show the figure
* `plt.savefig()` save the figure in the specific direction and name
    

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 6, 1000)
y = np.sin(3*x)
y2 = np.cos(3*x)

plt.figure(figsize = (4, 3))
plt.title("Tutorial Figure")

plt.plot(x, y, label='sin')
plt.plot(x, y2, label='cos')

plt.xlabel('x', fontsize = 12)
plt.ylabel('y', fontsize = 12)
plt.legend(loc='best')

plt.grid('--')
plt.show()
```

## Subplot

### Set the figure
* We can use `fig = plt.figure()` to set the figure
* We can use `(figsize=())` to set the size of the figure
* Using `fig.suptitle()` to set the title of the whole figure

### Set the axes
* `ax = plt.subplot()` to set the position of each axes
* There are some common set-up about setting axes
    1. `ax = set_title()` => Setting the title of each axes
    2. `ax = plot()` => Plot the line plot of each axes
    3. `ax = scatter()` => Plot the scatter plot of each axes
    4. `ax = set_xlabel()` => Set the label of x-axis of each axes
    5. `ax = set_ylabel()` => Set the label of y-axis of each axes
    6. `ax = legend(loc=)` => Set the position and the title of the legend of each axes
        * `loc='best'`
        * `loc='upper right'`
        * `loc='upper left'`
        * `loc='lower right'`
        * `loc='lower left'`
        * etc.
    7. `plt.tight_layout()` => Optimize the spacing between each subplot

### Show/Save the figure
* `plt.show()` show the figure
* `plt.savefig()` save the figure in the specific direction and name

```python
fig = plt.figure(figsize = (8, 3))
fig.suptitle("Total title")

x = np.linspace(0, 6, 1000)
y = np.sin(3*x)
ax1 = plt.subplot(121)
ax1.set_title('Sub title 1')
ax1.plot(x, y, label = 'sin')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend(loc='upper right')

x = np.linspace(0, 6, 1000)
y = np.cos(3*x)
ax2 = plt.subplot(122)
ax2.set_title('Sub title 2')
ax2.plot(x, y, label = 'cos')
ax2.set_yticks([])
ax2.set_xlabel('x')
ax2.legend(loc='upper right')

plt.tight_layout()

plt.show()
```
